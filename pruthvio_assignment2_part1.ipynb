```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sb
```

```python
# Reading the datasets
amazon=pd.read_csv('datasets/Amazon_top_selling_book.csv')
diamonds=pd.read_csv('datasets/diamonds.csv')
insurance=pd.read_csv('datasets/insurance.csv')
```

**1.Amazon Top Selling Book Dataset**

```python
# Working on the Amazon top selling book dataset

# Printing the shape of the dataset
print(f"The shape of the dataset is:\n {amazon.shape[0]} rows , {amazon.shape[1]} columns")
# Printing the description of the dataset.
print("Description \n",amazon.describe())
# Printing the heads of the datasets.
print("Heads \n",amazon.head()) 
```

```python
#Printing the memory used by each of the columns in the dataset
print("Memory Usage \n",amazon.memory_usage(deep=True))
#Printing the first 4 rows using .loc
print("\n First 4 rows",amazon.loc[0:4])
#checks whether the dataframe is empty or not
print("\n",amazon.empty)
```

```python
#Printing the total count of books released year wise
print("Total Count:\n",amazon['Year'].value_counts(),"\n") 
#Checking if there any null values in each of the columns
amazon.isnull().sum(axis=0)
```

**Data visulization of Amazon top selling book Dataset**

```python
#Plotting a graph for the user rating showing the count of highest number of rating
amazon['User Rating'].value_counts().plot(kind='bar')
plt.xlabel("rating")
plt.ylabel("count")
```

```python
#Plotting a pie graph to show the total number of books published in different years
amazon['Year'].value_counts().plot(kind='pie')
```

```python
#Plotting a bar graph to represent different types of genres
book_type = [0,0]
for i in amazon["Genre"]:
    if i=="Fiction":
        book_type[0] += 1
    elif i=="Non Fiction":
        book_type[1] += 1
    else:
        book_type[2] += 1
print (book_type)

x_axis=['Fiction','Non-Fiction','Others']
book_type_graph=go.Figure([go.Bar(x=x_axis, y=book_type)])
book_type_graph.show()
```

```python
#Plotting a scatter graph to show the variations of price with respect to year
fig = px.box(amazon,y="Author",x="Genre")
fig.show()
```

```python
#Plotting a histogram to show the number of books each other has published
fig = px.histogram(amazon, x="Author")
fig.show()
```
