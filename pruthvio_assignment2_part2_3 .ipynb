###NORMALIZATION OF DATA

```python
#normalization of the dataset
def normalize(x):
    x_norm = []
    x_max = max(x)
    x_min = min(x)
    for x_i in x:
        x_norm.append((x_i - x_min)/(x_max - x_min))
    return np.array(x_norm)
```

###LINEAR REGRESSION MODEL USING OLS

###splitting the data into testing and training dataset

```python
#initiazation of the target and variabel
x = df_penguin.iloc[:,:3]
y = df_penguin.iloc[:,[3]]
```

```python
#print(x)
```

```python
#defining a function to create test train split
def split_train_test(x, y):
    length = int(x.shape[0] * 0.8)
    x_train = x.iloc[:length, :]
    x_test = x.iloc[length:, :]
    y_train = y.iloc[:length]
    y_test = y.iloc[length:]
    return [np.array(x_train), np.array(x_test), np.array(y_train), np.array(y_test)]
```

```python
x_train, x_test, y_train, y_test = split_train_test(x, y)
```

```python
#printing shapes of test train split
print(x_train.shape,'\n')
print(x_test.shape,'\n')
print(y_train.shape,'\n')
print(y_test.shape,'\n')
```

```python
#creating a class for oridnary least square 
class LinearRegression_ols():
        
    def fit(self, x_train, y_train):
        train_ones = np.array(np.ones(shape = [x_train.shape[0], 1]))
        train_ones_new = np.resize(train_ones, (x_train.shape[0]))
        X_train = np.column_stack((train_ones_new, x_train))
        self.weights = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
        self.loss = self.cost(X_train, y_train, self.weights)
        return
    
    def cost(self, X_train, y_train, weights):
        return ((y_train - (X_train @ weights)).T @ (y_train - (X_train @ weights))) / y_train.shape[0] 
    
    def get_params(self):
        return [self.weights, self.loss]
    
    def predict(self, x_test):
        test_ones = np.array(np.ones(shape = [x_test.shape[0], 1]))
        test_ones_new = np.resize(test_ones, (x_test.shape[0]))
        X_test = np.column_stack((test_ones_new, x_test))
        return X_test @ self.weights
```

```python
#function mean square error
def mean_squared_error(y_pred, y_test):
    return ((y_test - y_pred).T @ (y_test - y_pred)) / y_test.shape[0]
```

```python
#training the model using old
model_ols = LinearRegression_ols()
model_ols.fit(x_train, y_train)
y_pred = model_ols.predict(x_test)
model_ols_params = model_ols.get_params()
print("Weights: \n", model_ols_params[0])
print("Loss: ", model_ols_params[1][0][0])
error = mean_squared_error(y_test, y_pred) 
print("Error: ", error[0][0])
```

### PART 3: RIDGE REGRESSION MODEL USING ORIDNARY LEAST SQUARE(OLS)

```python
class RidgeRegression():
    
    def __init__(self, lamdaa):
        self.lamdaa = lamdaa
        
    def fit(self, x_train, y_train):
        training_ones = np.array(np.ones(shape = [x_train.shape[0], 1]))
        training_ones_new = np.resize(training_ones, (x_train.shape[0]))
        X_train = np.column_stack((training_ones_new, x_train))
        I = np.identity(X_train.shape[1])
        self.weights = np.linalg.inv((X_train.T @ X_train) @ (self.lamdaa * I)) @ X_train.T @ y_train
        self.loss = self.cost(X_train, y_train, self.weights)
        return
    
    def cost(self, X_train, y_train, weights):
        return (((y_train - (X_train @ weights)).T @ (y_train - (X_train @ weights))) + (self.lamdaa * (weights.T @ weights))) / 2 
    
    def get_params(self):
        return [self.weights, self.loss]
    
    def predict(self, x_test):
        testing_ones = np.array(np.ones(shape = [x_test.shape[0], 1]))
        testing_ones_new = np.resize(testing_ones, (x_test.shape[0]))
        X_test = np.column_stack((testing_ones_new, x_test))
        return X_test @ self.weights
```

```python
class RidgeRegression_gsd():
    def __init__(self, learning_rate, iterations, lamdaa):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.lamdaa = lamdaa
    
    def fit(self, x_train, y_train):
        self.X = x_train
        self.Y = y_train
        np.random.seed(8)
        self.w = np.random.uniform(0, 1, self.X.shape[1])
        self.w = self.w.reshape([self.X.shape[1], 1])
        self.loss = []
        for i in range(self.iterations):
            self.loss.append(self.cost())
            self.gradient_descent()
        return self.loss, self.w
        
    def cost(self):
        self.J = 1/2*np.sum((self.Y-np.dot(self.X, self.w))**2) + 1/2*self.lamdaa*np.dot(self.w.T, self.w)
        return self.J.item()
        
    def gradient_descent(self):
        pred = np.dot(self.X, self.w)
        delta = pred - self.Y
        dw = 1/self.X.shape[0]*(np.dot(self.X.T, delta) + self.lamdaa*self.w)
        self.w = self.w - self.learning_rate * dw
        
    def predict(self, X):              
        return np.dot(X, self.w)
```
